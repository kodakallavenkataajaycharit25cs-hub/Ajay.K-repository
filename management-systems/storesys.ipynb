```python
# Smart Store Management System

inventory = {
    "Apple" : {"price" : 20 , "quantity" : 30},
    "Kiwi" : {"price" : 30 , "quantity" : 10},
    "Peanut Butter" : {"price" : 150 , "quantity" : 30},
    "Bread" : {"price" : 20 , "quantity" : 20},
    "Eggs" : {"price" : 10 , "quantity" : 50}
}

def add_product():
    name = input("Enter the name of item : ").capitalize()
    price = float(input("Enter the price : "))
    quantity = int(input("Enter the quantity : "))
    inventory[name] = {"price" : price , "quantity" : quantity}
    print(f"‚úÖ {name} successfully added")

def view_products():
    if not inventory:
        print("‚ùå Inventory is empty")

    else:
        print("\n üì¶üì¶ Products Available now")
        for name , info in inventory.items():
            print(f" - {name} : Price $ - {info['price']} , Quantity - {info['quantity']}")

def search_product():
    name = input('Enter the product name ').capitalize()

    if name in inventory:
        print("‚úÖ Found the product")
        info = inventory[name] 
        print(f"{name} : Price - {info['price']} , Quantity present - {info['quantity']}")

    else:
        print("‚ò†Ô∏è Product Not Found")

def update_inventory():
    name = input("Enter the name of product : ")
    if name in inventory:
        new_qty = int(input("Enter the new quantity : "))
        inventory[name]["quantity"] = new_qty
        print(f"üîÅ {name} quantity is updated to {new_qty}")

    else:
        print("üòë No product found")

def generate_bill():
    cart = []
    total = 0

    print("Enter product names to buy (type 'done' when finished)")
    while True:
        item = input(" > ").capitalize()
        if item.lower() == 'done':
            break

        if item in  inventory:
            if inventory[item]["quantity"] > 0:
                cart.append(item)
                total += inventory[item]['price']
                inventory[item]['quantity'] -= 1
            else: 
                print("‚ö†Ô∏è Out Of Stock")
        else:
            print("‚ùå Product Not Found")

    print("\n üìÉ Generating E-bill ...")
    for item in cart:
        print(f" {item}: ${inventory[item]['price']}") 
    print(f"Total Amount : ${total}")
    print("üôá‚Äç‚ôÇÔ∏è Thanks for Shopping with Us!")

while True:
    print("\n ---Store Management System---")
    print("1.Add Product")
    print("2.View Products")
    print("3.Search for a Product")
    print("4.Update Inventory")
    print("5.Generate E-Bill")
    print("6.Exit")
    choice = input("Enter your choice(1-6)")
    if choice == "1":
        add_product()

    elif choice == "2": 
        view_products()

    elif choice == "3": 
        search_product()

    elif choice == "4": 
        update_inventory()

    elif choice == "5": 
        generate_bill()

    elif choice == "6": 
        print("Exiting the System ...")
        print("Good Bye üëã")
        break 
```
