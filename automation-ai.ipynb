```python
import random
import time
import matplotlib.pyplot as plt
import IPython.display as display
import IPython

class MotorDigitalTwin:
    def __init__(self , motor_id):

        self.motor_id = motor_id
        self.temperature = 30 #starting temp
        self.vibrations = 0.4 #starting g force

        self.status = 'Normal'
    def update_sensors(self):
        self.temperature += 0.5 + random.uniform(-0.2 , 0.3)
        if random.random() < 0.1:
            self.vibrations += random.uniform(0.5,1.5)

        else:
            self.vibrations = max(0.2 , self.vibrations + random.uniform(-0.05 , 0.05))

    def analyze(self):
        if self.temperature > 65 or self.vibrations > 2.25:
            self.status = 'Failure Predictied ☠️☠️'

        elif self.temperature > 40 or self.vibrations > 1.5:
            self.status = 'Warning 🥵🥵'
        else:
            self.status = 'Normal 😊😊'
    def get_status(self):
        return {
            "Motor ID" : self.motor_id,
            "Temperature (C*)": round(self.temperature , 2),
            "Vibrations (g)" : round(self.vibrations , 2),

            'Status' :self.status
        }

motor =  MotorDigitalTwin("Motor-001")

temps = []
vibs = []
stats = []

plt.ion()
fig , ax1 = plt.subplots()

for cycle in range(30):
    motor.update_sensors()
    motor.analyze()
    status = motor.get_status()
    print(status)

    temps.append(status["Temperature (C*)"])
    vibs.append(status["Vibrations (g)"])
    stats.append(status['Status'])

    ax1.clear()
    ax1.set_title("Digital Twin : Motor Monitoring")
    ax1.set_xlabel("Cycles")
    ax1.set_ylabel("Temperature (C*)" , color = 'red')
    ax1.plot(temps , color = 'red' , label = "Temperature")
    ax1.tick_params(axis = 'y' , labelcolor = 'red')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Vibration (G)' , color = 'blue')
    ax2.plot(vibs , color = 'blue' , label = "Vibration")
    ax2.tick_params(axis = 'y' , labelcolor = 'blue')

    plt.pause(0.5)
    plt.show()
    IPython.display.display(fig)
    plt.close(fig)
```
