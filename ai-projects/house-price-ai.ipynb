```python
from sklearn.linear_model import LinearRegression

# x =  [(area)sq feet  ,  no of bedrooms , age]
x = [
    [1000 , 2 , 5],
    [700 , 1 , 7],   
    [1500, 3 ,10],
    [800 , 2 , 2]
]

y = [80, 50 , 95 , 67]

model = LinearRegression()
model.fit(x,y)

new_house = [[2000 , 4 ,7]]

prediction = model.predict(new_house)

print(f"The predicted price for the house is {prediction[0]: .2f} lakhs")

```
