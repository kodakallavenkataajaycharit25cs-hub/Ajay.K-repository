```python
# Library Mangement System
books = {
    "Atomic Habits" : {'Author' : 'James Clear' , "copies" : 4},
    "The Art of Being Alone" : { 'Author' : "Renuka Joshi" , "copies" : 6},
    "Ikagi" : { 'Author' : "Herman" , "copies" : 2},
    "Wings of fire" : {'Author' : 'Abdul Kalam' , "copies" : 1}
}

while True:
    print("\n ___Library Mangement System___ ")
    print("1.Add Book")
    print("2.Borow Book")
    print("3.Return Book")
    print("4.Display All Books")
    print("5.Exit")

    choice = input("Enter the requried action(1-5) : " )

    if choice == "1":
        title = input("Enter the name of book: ")
        author = input("Enter the Author name: ")
        copies = int(input("Enter the no of copies adding: "))

        books[title] = {"Author" : author , "copies" : copies}

        print(f"✅ '{title}' sucessfully added!")

    elif choice == "2":
        title = input("Enter the name of the book you are bowrowing: ")

        if title in books and books[title]["copies"] > 0:
            books[title]["copies"] -= 1
            print(f"You Borrowed '{title}'.")
        elif title in books:
            print("😔 Sorry copies are unavailable rigth at this moment")
        else:
            print("❌ We could'nt a book with that title")

    elif choice == "3":
        title = input("Enter the title of the book you are returning")

        if title in books:
            books[title]["copies"] += 1

            print("Book successfully returned 😊")

        else:
            print("No Book with that tile belongs to this library 😑")

    elif choice == "4":
        print("\n The Present Library Collection 📚")
        for title , info in books.items():
            print(f" . {title} - Author: {info['Author']} , Copies Present : {info['copies']}")


    elif choice == "5":
        print("Exiting Library Mangement System...")
        print("GoodBye 👋")
        break


    else:
        print("I dont understand that try giving serial no from (1 to 5)")
        
        
```
